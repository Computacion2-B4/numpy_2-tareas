### Ejercicio 1 - Investigar Como?

#### Resolver las siguientes ecuaciones:


$$12x + 5y + z= 6$$
$$6x + 10y + z = 8$$
$$3x + 15y + z = 4$$



#### Utilizando un método diferente al enseñado  $A*x =B$

```python
import numpy as np
import numpy.linalg as lin

A= np.array([
    [12,5,1],
    [6,10,1],
    [3,15,1]
])

B= np.array([
    [6],
    [8],
    [4]
])

#===========================================
x = lin.solve(A, B)

print('La matriz A es: ')
print(A)
print('\nLa matriz B es: ')
print(B)
print('\nLa matriz x , que sale al dividir la matriz B entre A, es: ')
print(x)
```

### Ejercicio 2 - Investigar Como?

Integrar **Ecuaciones diferenciales ordinarias** usar la función `odeint` del paquete `integrate`, que permite integrar sistemas del tipo:

$$ \frac{d\mathbf{y}}{dt}=\mathbf{f}\left(\mathbf{y},t\right)$$

con condiciones iniciales $\mathbf{y}(\mathbf{0}) = \mathbf{y_0}$.

```python
# Pista utilizar.......
from scipy.integrate import odeint
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define la función f(y, t)
def f(y, t):
    return -y  # Definimos la ecuación diferencial dy/dt = -y

# Define las condiciones iniciales y el tiempo de integración
y0 = 1.0                     # Condición inicial: y(0)
t = np.linspace(0, 10, 100)  # Vector de tiempo de 0 a 10 con 100 puntos

# Resuelve la EDO
y = odeint(f, y0, t)         # Resolvemos la EDO utilizando odeint

# Grafica la solución
plt.plot(t, y)               # Graficamos la solución en función del tiempo
plt.xlabel('Tiempo')         # Nombramos el eje x como tiempo
plt.ylabel('y(t)')           # Nombramos el eje y como y(t)
plt.title('Solución de la EDO dy/dt = -y')  # Título del gráfico
plt.grid(True)               # Ponemos una cuadricula en el fondo
plt.show()                   # Mostramos el gráfico


```

<div class="well">Realizar un ajuste polinomico en numpy del archivo <code>covid.txt</code> (archivo adjunto)
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import pandas as pd
```

```python
# Cargar datos desde el archivo covid.txt
datos = np.loadtxt('/content/drive/MyDrive/covid.txt')

# Generar los índices para los datos
indices = np.arange(len(datos))

# Realizar un ajuste polinomial de grado 5 a los datos
coeficientes = np.polyfit(indices, datos, 5)

# Crear un polinomio con los coeficientes obtenidos
ajuste_polinomial = np.poly1d(coeficientes)

# Generar datos para la línea de ajuste
x_fit = np.linspace(0, len(datos), 100)
y_fit = ajuste_polinomial(x_fit)

# Graficar los datos y el ajuste polinomial
plt.plot(indices, datos, 'o', label='Datos')
plt.plot(x_fit, y_fit, label='Ajuste polinomial (grado 5)')
plt.xlabel('Días')                              #Nombramos al eje x  como Dias
plt.ylabel('Casos de COVID-19')                 #Nombramos al eje y como Casos de COVID-19
plt.title('Ajuste Polinomial de Casos de COVID-19')    #Colocamos el titulo de la gráfica
plt.legend()                        #Ponemos la legenda (Datos y el ahuste polinomial)
plt.grid(True)                      #Ponemos la cuadricula en el fondo
plt.show()                          # Mostramos el gráfico
```

### Ejercicio 4 - Investigar Como?

<div class="well">Crear un programa que permita hacer operaciones con matrices en <strong>Python</strong> utilizando los módulos <strong>NumPy</strong> y <strong>SciPy</strong> para resolver la inversa, la multiplicación y la solución de un sistema de ecuaciones lineales.


</div>


```python
import numpy as np
import scipy.linalg as la

# Definir una matriz
A = np.array([[1, 2], [3, 4]])

# Inversa de la matriz
A_inv = la.inv(A)

# Multiplicación de matrices
B = np.array([[5, 6], [7, 8]])
C = np.dot(A, B)

# Resolver un sistema de ecuaciones lineales
# Ax = b
b = np.array([1, 2])
x = la.solve(A, b)

# Imprimir resultados
print("Matriz A:")
print('=========')
print(A)
print("\nInversa de A:")
print('===============')
print(A_inv)
print("\nMultiplicación de A y B:")
print('===========================')
print(C)
print("\nSolución del sistema de ecuaciones lineales Ax = b:")
print('=====================================================')
print(x)
```

### Ejercicio 5 - Investigar Como?

<div class="well">Crear un programa que permita resolver una <strong>interpolación utilizando polinomios de Lagrange</strong> en Python <strong> SciPy </strong> .


</div>

```python
from scipy.interpolate import lagrange

def interpolacion_lagrange(x, y, x_eval):
    """
    Realiza una interpolación utilizando polinomios de Lagrange.

    Args:
        x (list): Lista de puntos x conocidos.
        y (list): Lista de puntos y conocidos.
        x_eval (float): Valor x en el que se desea evaluar el polinomio interpolante.

    Returns:
        float: Valor y obtenido mediante interpolación en x_eval.
    """

    # Calcula el polinomio de Lagrange que pasa por los puntos (x, y)
    polinomio = lagrange(x, y)
    # Evalúa el polinomio de Lagrange en el valor dado x_eval
    y_eval = polinomio(x_eval)
    return y_eval

# Ejemplo de uso
x = [1, 2, 3, 4, 5]
y = [6, 4, 3, 2, 1]
x_eval = 2.5

resultado = interpolacion_lagrange(x, y, x_eval)
print('\n---------Conclusión----------')
print("El valor interpolado en x =", x_eval, "es y =", resultado)
```
